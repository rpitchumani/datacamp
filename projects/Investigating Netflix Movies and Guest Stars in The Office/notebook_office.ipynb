```python
# Use this cell to begin your analysis, and add as many as you would like!
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./datasets_office/office_episodes.csv")
df
```

```python
df.info()
```

```python
conditions_color = [
    (df["scaled_ratings"] < 0.25),
    (df["scaled_ratings"] >= 0.25) & (df["scaled_ratings"] < 0.50),
    (df["scaled_ratings"] >= 0.50) & (df["scaled_ratings"] < 0.75),
    (df["scaled_ratings"] >= 0.75),
]
choices_color = ["red", "orange", "lightgreen", "darkgreen"]

conditions_marker_size = [
    (df["has_guests"]==True),
    (df["has_guests"]==False),
]
choices_marker_size = [250, 25]

df["color"] = np.select(conditions_color, choices_color)
df["marker_size"] = np.select(conditions_marker_size, choices_marker_size)

df.head()
```

```python
fig = plt.figure(figsize=[8, 6])

plt.scatter(
    x=df["episode_number"],
    y=df["viewership_mil"],
    c=df["color"],
    s=df["marker_size"]
)

plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.show()
```
